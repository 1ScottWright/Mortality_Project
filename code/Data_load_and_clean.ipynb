# Compute rate of change in mortality given a vector accross time and plot results.  Expecting to see "negative improvement" in recent years, particularly for males  

#### This is just for EDA.  More sophisticated functions will need to be compiled later.

```python
def plot_rate_of_change(qx_vector):
    qx_vector = np.array(qx_vector)
    j = len(qx_vector)
    if j <= 1:
        print("Mortality info passed to function not large enough to compute qx improvement.")
        pass
    qx_improve = []
    for i in range(j-1):
        qx_improve.append(1-qx_vector[i+1]/qx_vector[i])

    qxi = pd.DataFrame({'improvement':qx_improve})
    qxi['improvement'].plot(kind='line')
    plt.show();
    print(qxi)
    return
    
    
```

```python
plot_rate_of_change(male.iloc[35,-20:])
```

```python
plot_rate_of_change(female.iloc[35,-20:])
```

```python
plot_rate_of_change(male.iloc[55,-20:])
```

```python
plot_rate_of_change(female.iloc[55,-20:])
```

```python
plot_rate_of_change(male.iloc[70,-20:])
```

```python
plot_rate_of_change(female.iloc[70,-20:])
```
